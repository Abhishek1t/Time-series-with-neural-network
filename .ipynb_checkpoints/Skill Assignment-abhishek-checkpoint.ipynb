# Data Transformation

```python
data["Interval"]=data["Mbps"]
m=1
for i in range(1,4033):
    if i%96!=0:
        data.at[i-1,"Interval"]=m
        m=m+1
    else:
        data.at[i-1,"Interval"]=96
        m=1
```

Introducing a new column(Interval) wich will represent the 15 min interval of the day. 

```python
l=list(data["Mbps"][1:])
l.append(data["Mbps"][0])
data["y"]=l
```

This new column "y" will be our target variable which is the usage of the data in the next interval

```python
df=data[["Interval","Mbps"]].groupby(["Interval"],as_index=False).mean().sort_values(by="Interval")
plt.figure(figsize=(14,5))
plt.plot(df)
plt.title("Usage Through out the day")
plt.xlabel("Each 1 unit is of 15 min")
plt.ylabel("Usage of Mbps through the day")
plt.show()
```

We can observe from the above graph that the lowest usage of internet is at 5'O clock in the morning. 
The usage of internet start at a high peak at the start of the day and the usage starts to reduce till it reaches 5'O clock in the morning which is the lowest usage of the whole day. Then it again start to rise and peaks at the end of the day.

### Splitting Data into train-test split

```python
x_train = data[["Interval","Mbps"]][:3226]
x_test  = data[["Interval","Mbps"]][3226:]
y_train = data["y"][:3226]
y_test  = data["y"][3226:]
x_train.shape ,x_test.shape , y_train.shape ,y_test.shape
```

```python
from keras import models, layers
model = models.Sequential()
model.add(layers.Dense(32,activation="relu"))
model.add(layers.Dense(64,activation="relu"))
model.add(layers.Dense(1,activation="linear"))

#compile model
model.compile(loss="mean_absolute_error",optimizer="adam",metrics=["mse","mae"])

#fit the model
model.fit(x_train, y_train, validation_split = 0.3, epochs=20, batch_size=1)

test_acc, test_loss, test = model.evaluate(x_test, y_test)
print('Test Accuracy : ', test_acc)
```

```python
y_pred = model.predict(x_test)
print("Predicted :""\n",y_pred[0:5],"\n""Actual :",list(y_test[0:5]))
```

### Plotting the Prediction vs Actual usage graph

```python
plt.figure(figsize=(17,4))
# prediction usage graph  
plt.subplot(1,2,1)
plt.title("Prediction of Mbps usage")
plt.plot(y_pred,color = "blue")
plt.xlabel("Interval")
plt.ylabel("Mbps Usage")
#actual usage graph
plt.subplot(1,2,2)
plt.title("Actual Mbps usage")
plt.plot(list(y_test),label="actual",color = "green")
plt.xlabel("Interval")
plt.ylabel("Mbps Usage")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(17,5))
plt.title("Actual usage vs Predicted usage")
plt.plot(y_pred,label="Predicted usage",color = "blue")
plt.plot(list(y_test),label="Actual usage",color = "green")
plt.xlabel("Observation Interval")
plt.ylabel("Mbps usage")
plt.legend()
plt.show()
```
